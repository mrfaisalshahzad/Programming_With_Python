```python
# Conditional operator (ternary operator) in Python
# Syntax: value_if_true if condition else value_if_false

# List of conditional operators and thier names:
# == : Equal to
# != : Not equal to
# > : Greater than
# < : Less than
# >= : Greater than or equal to
# <= : Less than or equal to

x = int(input("What's x? "))
y = int(input("What's y? "))

if x < y:
    print("x is less than y")
if x > y:
    print("x is greater than y")
if x == y:
    print("x is equal to y")
```

```python
name = input("Whats's your name? ")

match name:
    case "Faisal":
        print("Hello Faisal")
    case "Alice":
        print("Hello Alice")
    case "Bob":
        print("Hello Bob")
    case _:
        print("Who are you?")
```

```python

# Dictionary: Best for representing a single "User" or "Role" record: {"user_id": 1, "role": "Admin"}.
# List: Best for holding multiple records: [{"user_id": 1...}, {"user_id": 2...}].
# Tuple: Best for fixed constants, like a list of permission levels that should never change during runtime: PERMISSIONS = ("READ", "WRITE", "DELETE").

def main():
    user_input = input("File name: ").strip().lower()

    ext_types = {
        ".gif": "image/gif",
        ".jpg": "image/jpeg",
        ".jpeg": "image/jpeg",
        ".png": "image/png",
        ".pdf": "application/pdf",
        ".zip": "application/zip",
        ".txt": "text/plain"
    }

    ext_start = user_input.rfind(".")
    ext = user_input[ext_start:]

    if ext in ext_types.keys():
        reply(ext_types.get(ext))
    else:
        reply("application/octet-stream")
    

def reply(answer):
    print(answer)

main()
```

```python
def main():
    time = input("What time is it? ").strip()

    
    meal_time = convert(time)

    if 7 <= meal_time <= 8:
        print("breakfast time")
    elif 12 <= meal_time <= 13:
        print("lunch time")
    elif 18 <= meal_time <= 19:
        print("dinner time")


def convert(time):
    # hr, mnt = time.split(":")
    hr = float(time.split(":")[0])
    mnt = float(time.split(":")[-1])/60
    return hr + mnt


if __name__ == "__main__":
    main()
```
