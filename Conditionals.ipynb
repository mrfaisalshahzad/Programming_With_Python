```python
# Conditional operator (ternary operator) in Python
# Syntax: value_if_true if condition else value_if_false

# List of conditional operators and thier names:
# == : Equal to
# != : Not equal to
# > : Greater than
# < : Less than
# >= : Greater than or equal to
# <= : Less than or equal to

x = int(input("What's x? "))
y = int(input("What's y? "))

if x < y:
    print("x is less than y")
if x > y:
    print("x is greater than y")
if x == y:
    print("x is equal to y")
```

```python
# Use of the modulo operator in conditionals

x = int(input("What's x? "))

if x % 2 == 0:
    print("x is even")
else:
    print("x is odd")
```

```python
# Using match-case statement

name = input("Whats's your name? ")
if name == "Faisal":
    print("Hello Faisal")
elif name == "Alice":
    print("Hello Alice")
elif name == "Bob":
    print("Hello Bob")
else:
    print("Who are you?")
```

```python
name = input("Whats's your name? ")
if name == "Faisal" or name == "Alice" or name == "Bob":
    print("Hello " + name)
else:
    print("Who are you?")
```

```python
name = input("Whats's your name? ")

match name:
    case "Faisal":
        print("Hello Faisal")
    case "Alice":
        print("Hello Alice")
    case "Bob":
        print("Hello Bob")
    case _:
        print("Who are you?")
```

```python
name = input("Whats's your name? ")

match name:
    case "Faisal" | "Alice" | "Bob":
        print("Hello " + name)
    case _:
        print("Who are you?")
```
