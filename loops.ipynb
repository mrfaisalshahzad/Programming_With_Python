```python
# While Loops

i = 0
while i < 3:
    print("meow")
    i += 1
    print(i)
```

```python
# For Loops

for i in [0, 1, 2]:
    print("meow")
    print(i)
```

```python
for _ in range(3): # instead of i we can also use _ if we don't care about this variable, because you are not using it later.
    print("meow")
```

```python
print("meow\n" *3, end = "")
```

```python
while True:
    n = int(input("What's n? "))
    if n > 0:
        break

for _ in range(n):
    print("meow")
```

```python
def main():
    number = get_number()
    meow(number)

def get_number():
    while True:
        n = int(input("What's n? "))
        if n > 0:
            break
    return n

def meow(n):
    for _ in range(n):
        print("meow")

main()
```

```python
# Lists

students = ["faisal", "adil", "arslan"]

for student in students:
    print(student)
```

```python
students = ["faisal", "adil", "arslan"]

for i in range(len(students)):
    print("idx", i, students[i])
```

```python
# Dictionaries

students = {"faisal": "house1",
          "adil": "house2",
          "arslan": "house3" 
}

print(students["faisal"])
print(students["arslan"])
```

```python
students = {"faisal": "house1",
          "adil": "house2",
          "arslan": "house3" 
}

for student in students:
    print(student, students[student], sep = ", ")
```
