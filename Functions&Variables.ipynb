```python
# print function syntex: print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)
# explanation of the print function syntex:
# *objects: the objects to be printed, separated by sep
# sep: the separator between the objects, default is a space
# end: the string appended after the last object, default is a newline
# file: the file to write the output to, default is sys.stdout
# flush: whether to forcibly flush the stream, default is False

# objects, sep and end are also called named peramaters, they can be used in any order and they can be omitted if they are not needed


name = input("What's your name? ")
print("hello", name, sep=" ", end=", ")
print("welcome to the world of programming with python!")
```

```python
name = input("What's your name? ")
print('hello, "Faisal"')
print("hello, \"faisal\"")

# f-string is a string literal that is prefixed with 'f' or 'F' and contains expressions inside curly braces {} that are evaluated at runtime and formatted using the __format__ protocol
# f-string is a format string

print(f"hello, {name}")
```

```python
name = input("what's your name? ")
print("hello, " , name)

# Remove whitespace from the beginning and end of the string
# Where .strip() is a method that removes any leading and trailing whitespace from a string. It does not modify the original string, but returns a new string with the whitespace removed.
name = name.strip()

# Capitalize the first letter of the string and make the rest of the letters lowercase
name = name.capitalize().title()

# name = input("what's your name? ").strip().capitalize().title() is the same as the above code but in one line

# Split user's name into first name and last name
first, last = name.split(" ")
print(f"hello, {first}")

print("hello," , name)
print(name)
```
