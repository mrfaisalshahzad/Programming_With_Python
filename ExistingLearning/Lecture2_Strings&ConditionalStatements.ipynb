**Strings**

Strings is data type that stores a sequence of characters.
Strings are immutable

```python
str1 = "this is a string"
str2 = 'Faisal Shahzad'
str3 = """this is a string"""
str4 = "Faisal's age is 38 years"

print(str1)
print(str2)
print(str3)
print(str4)
```

**Escape Sequence Characters in Strings**

- `\\` : Backslash
- `\'` : Single Quote  
- `\"` : Double Quote  
- `\n` : New Line  
- `\t` : Tab  
- `\r` : Carriage Return  
- `\b` : Backspace  
- `\f` : Form Feed  
- `\v` : Vertical Tab  
- `\a` : Bell/Alert  
- `\N{name}` : Unicode Character by Name  
- `\uXXXX` : Unicode Character with 16-bit Hex Value  
- `\UXXXXXXXX` : Unicode Character with 32-bit Hex Value  
- `\xhh` : Character with Hex Value hh  
- `\ooo` : Character with Octal Value ooo

```python
# use of escape sequence characters
str5 = 'Faisal\'s age is 38 years'
str6 = "This is line 1.\nThis is line 2."
str7 = "Faisal said, \"Hello, how are you?\""
str8 = "This is a backslash: \\"
str9 = "This is a tab:\tTabbed text"
str10 = "Hello World!\rHi"
str11 = "Hello World!\b"
str12 = "Hello World!\f"
str13 = "Hello World!\v"
str14 = "This is a Unicode character: \u03A9"  # Omega symbol
str15 = "This is a Unicode character: \U0001F600"  # Grinning Face
str16 = r"This is a raw string: C:\Users\Faisal"

print(str5)
print(str6)
print(str7)
print(str8)
print(str9)
print(str10)
print(str11)
print(str12)
print(str13)
print(str14)
print(str15)
print(str16)
```

```python
# String concatenation

str1 = "Hello"
str2 = "World"
str3 = str1 + " " + str2
print(str3)
```

```python
# Length of strings

print(len(str1))
print(len(str2))
print(len(str3))
```

**Slicing Strings**

Accessing parts of a string

```python
str = "Hello, World!"
print(str[1:5])
print(str[0:7])
print(str[1:])
print(str[:8])
print(str[:])
print(str[::2]) # step size
print(str[::-1]) # reverse a string
print(str[:len(str)])

# Negative indexing

print(str[-3:-1]) # last character is not included
print(str[-5:len(str)])
```

**String Functions**

- `len()` – Returns the length of the string  
- `str.lower()` – Converts all characters to lowercase  
- `str.upper()` – Converts all characters to uppercase  
- `str.capitalize()` – Capitalizes the first character  
- `str.title()` – Capitalizes the first character of each word  
- `str.strip()` – Removes leading and trailing whitespace  
- `str.lstrip()` – Removes leading whitespace  
- `str.rstrip()` – Removes trailing whitespace  
- `str.replace(old, new)` – Replaces a substring with another  
- `str.split(sep)` – Splits the string into a list  
- `str.join(iterable)` – Joins elements of an iterable with the string  
- `str.find(sub)` – Returns the lowest index of the substring  
- `str.rfind(sub)` – Returns the highest index of the substring  
- `str.count(sub)` – Counts occurrences of a substring  
- `str.startswith(prefix)` – Checks if string starts with prefix  
- `str.endswith(suffix)` – Checks if string ends with suffix  
- `str.isalpha()` – Checks if all characters are alphabetic  
- `str.isdigit()` – Checks if all characters are digits  
- `str.isalnum()` – Checks if all characters are alphanumeric  
- `str.isspace()` – Checks if all characters are whitespace  
- `str.swapcase()` – Swaps case of all characters  
- `str.zfill(width)` – Pads string on the left with zeros  
- `str.center(width)` – Centers string in a field of given width  
- `str.ljust(width)` – Left justifies string  
- `str.rjust(width)` – Right justifies string  
- `str.partition(sep)` – Splits string into a tuple (before, sep, after)  
- `str.rpartition(sep)` – Splits string into a tuple from the right  
- `str.encode()` – Encodes string into bytes  
- `str.format()` – Formats string using placeholders

```python
str = "hello, world!"
print(str.endswith("rld!"))
print(str.capitalize()) #does not change the original string
#str = str.capitalize() # to change the original string
print(str)

str = "hello, world!"
print(str.replace("o","A"))

str = "Hello, World!"
print(str.find("o"))
print(str.find("q")) # returns -1

str = "Hello, World!"
print(str.count("l"))
```

**Conditional Statements**

if-elif-else (SYNTAX)

```python
age = 18

if(age >= 18):
    print("Can vote and apply for a license.")
    print("Can drive")

light = "yellow"

if(light == "green"):
    print("Go")
elif(light == "yellow"):
    print("ready to stop")
elif(light == "red"):
    print("stop")
```

```python
num = 5

if(num > 2):
    print("greater than 2")
elif(num > 5):
    print("greater than 5")
else:
    print("less than or equal to 2")
```

```python
marks = float(input("Enter student's marks: "))

if(marks >= 90):
    print("Grade A")
elif(marks >= 80 and marks < 90):
    print("Grade B")
elif(marks >= 70 and marks < 80):
    print("Grade C")
else:
    print("Grade D")
```

```python
age = 80

if(age >= 18):
    if(age >= 80):
        print("Can't drive")
    else:
        print("Can drive")
else:
    print("Can't drive")
```
