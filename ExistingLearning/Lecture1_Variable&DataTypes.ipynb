**Types of Operators**

An operator is a symbol that performs a certain operation between operands.

* Arithmatic operators (+, -, *, /, %, **)
* Relational / comparison Operators (==, !=, >, >=, <, <=)
* Assignment Operators (=, +=, -=, *=, /=, %=, **=)
* Logical Operators (not, and , or)

```python
# arithmetic operators
a = 5
b = 2
c = a + b
d = a - b
e = a * b
f = a / b
g = a % b # modulus operator (remainder)
h = a ** b # exponentiation operator or power operator
i = a // b # floor division operator

print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
print(i)
```

```python
#relational operators

a = 50
b = 20
c = a == b #False
d = a != b #True
e = a > b #True
f = a < b #False
g = a >= b #True
h = a <= b #False

print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
```

```python
#assignment operators

num = 10
num += 5 # equivalent to num = num + 5
num -= 3 # equivalent to num = num - 3
num *= 2 # equivalent to num = num * 2
num /= 4 # equivalent to num = num / 4
num %= 3 # equivalent to num = num % 3
num **= 2 # equivalent to num = num ** 2
num //= 2 # equivalent to num = num // 2

print(num)
```

```python
#logical operators

a = True
b = False
c = a and b #False - both must be true
d = a or b #True - at least one must be true
e = not a #False - negates the value

print(c)
print(d)
print(e)
```

**Type Conversion**

* Type conversion (phthon does this automaticaly also called **Implicit conversion**)
* Type casting (it is done manually/intentionaly also called **Explicit conversion**)

```python
#type conversion

a = 5
b = 2.5
c = "10"

# Implicit conversion
d = a + b  # a is int, b is float, result is float

# Explicit conversion
e = a + int(c)  # c is converted to int

print(d)
print(e)
```
