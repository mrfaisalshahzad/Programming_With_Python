**List in Python**

[]
A built in data type that stores a set of values
It can store elements of different types (int, float, str, etc.)
Lists are mutable

```python
marks = ["Faisal", 78.9, "Adil", 88.0, "Arslan", 92.5]
print(marks)
print(type(marks))
print(marks[0])
print(marks[1])
print(len(marks))
print(marks[-1])

marks[2] = "ADIL" #lists are mutable
print(marks)
```

```python
# list slicing

marks = ["Faisal", 78.9, "Adil", 88.0, "Arslan", 92.5]
print(marks[1:4])
print(marks[:4])
print(marks[2:])
print(marks[:])
print(marks[::2]) # step size
print(marks[::-1]) # reverse a list
print(marks[-1:-4:-1]) # reverse a part of list
print(marks[-4:-1])
print(marks[-1:-4]) # empty list

```

**List Methods**

- `append(x)` – Add an item to the end of the list  
- `extend(iterable)` – Extend the list by appending elements from the iterable  
- `insert(i, x)` – Insert an item at a given position  
- `remove(x)` – Remove the first item with the value x  
- `pop([i])` – Remove and return item at position i (default last)  
- `clear()` – Remove all items from the list  
- `index(x[, start[, end]])` – Return first index of value x  
- `count(x)` – Return the number of times x appears  
- `sort(key=None, reverse=False)` – Sort the list in place  
- `reverse()` – Reverse the elements of the list in place  
- `copy()` – Return a shallow copy of the list

```python
marks = ["Faisal", 78.9, "Adil", 88.0, "Arslan", 92.5]
marks.append("Ahmed")
print(marks.append("Ahmed")) # returns None
marks.extend(["Ahmed2", 85.0])  # Adds both "Ahmed" and 85.0 to the list
print(marks)
```

```python
marks = [90, 85, 88, 92, 95, 80, 87]
marks.sort() # sorts the list in ascending order
print(marks.append(100)) # returns None
marks.sort(reverse=True) # sorts the list in descending order
print(marks)
```

```python
fruits = [ "date", "banana", "apple", "cherry", "elderberry"]
print(fruits)
fruits.sort() # sorts the list in ascending order
print(fruits)
fruits.sort(reverse=True) # sorts the list in descending order
print(fruits)
```

```python
marks = [90, 85, 88, 88, 95, 80, 87]
marks.insert(1,91) # inserts 91 at index 1
print(marks)
marks.remove(88) # removes first occurrence of 88
print(marks)
marks.pop(2) # removes and returns element at index 2
print(marks)
```

**Tuples in Python**

()
A built-in data type that lets us create immutable sequences of values.

```python
tup = (90, 85, 88, 88, 95, 80, 87,) # tuple with trailing comma
print(tup)
print(type(tup))
print(tup[0])

tup = () # empty tuple
print(tup)
print(type(tup))

tup = (90,) # single element tuple
print(tup)
print(type(tup))
print(tup[0])

tup = (1) # not a tuple
print(tup)
print(type(tup)) # int, not tuple
```

```python
# tuple slicing

tup = (90, 85, 88, 88, 95, 80, 87,) # tuple with trailing comma
print(tup[0:3])
print(tup[3:])
print(tup[:4])
print(tup[:])
print(tup[::2]) # step size
```

**Tuple Methods**

- `count(x)` – Return the number of times x appears in the tuple  
- `index(x[, start[, end]])` – Return first index of value x

```python
tup = (90, 85, 88, 88, 95, 80, 87,) # tuple with trailing comma
print(tup.index(88)) # returns index of first occurrence of 88
print(tup.count(80)) # returns count of occurrences of 88
```

```python
# palindrome list

list1 = [1, 2, 3, 4, 5]
list2 = [5, 4, 3, 2, 1]

copy_list1 = list1.copy()
copy_list1.reverse()

if(copy_list1 == list2):
    print("list if palindrome")
else:
    print("list is not palindrome")
```
